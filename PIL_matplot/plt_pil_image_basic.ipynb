# matplotlib

```python
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import numpy as np
```

### show image

```python
img = mpimg.imread('img/Grass.JPG') 

print(img.shape)
plt.imshow(img) 
plt.axis('off')
plt.show()
```

### show image channel

```python
img_1 = img[:,:,0]
plt.imshow(img_1)
plt.show()
```

```python
plt.imshow(img_1, cmap='Greys_r')
plt.show()
```

```python
img = plt.imshow(img_1)
img.set_cmap('gray') 
plt.show()
```

### rgb 2 gray

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

img = mpimg.imread('img/Grass.JPG') 
gray = rgb2gray(img)    
plt.imshow(gray, cmap='Greys_r')
plt.axis('off')
plt.show()
```

### image resize

```python
from scipy import misc

img = mpimg.imread('img/Grass.JPG') 
img_new = misc.imresize(img, 0.5, 'nearest') 
plt.imshow(img_new)
plt.axis('off')
plt.show()
```

### image save

```python
#plt.savefig('img/grass_new.png')


from scipy import misc
misc.imsave('img/grass_new2.png', img_new)

np.save('img/grass_new', img_new)

# read image
img = np.load('img/grass_new.npy') 
plt.imshow(img) 
#plt.axis('off')
plt.show()
```
